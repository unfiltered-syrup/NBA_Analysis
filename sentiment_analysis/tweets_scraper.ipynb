```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time
import random
```

```python
def remove_ads(driver):
    all_iframes = driver.findElement(By.Tagname("iframe"))
    if len(all_iframes) > 0:
        print("Ad Found\n")
        driver.execute_script("""
            var elems = document.getElementsByTagName("iframe"); 
            for(var i = 0, max = elems.length; i < max; i++)
                {
                    elems[i].hidden=true;
                }
                            """)
        print('Total Ads: ' + str(len(all_iframes)))
    else:
        print('No frames found')
```

```python
teams = ['golden_state_warriors', 'houston_rockets', 'los_angeles_clippers', 'los_angeles_lakers', 'utah_jazz', 'charlotte_hornets']
#sets how many times a page is scrolled
for team in teams:
    n_iters = 180
    scroll_time = 8
    service = Service() #initialize the service
    options = webdriver.ChromeOptions()
    driver = webdriver.Chrome(service=service, options=options)
    url = "https://hoopshype.com/social/tweets/"+team+"/"
    driver.get(url)
    unchanged_iter_count = 0
    #scroll the page to load more tweets
    for i in range(n_iters):
        print('current iteration:', i)
        last_height = driver.execute_script("return document.body.scrollHeight")
        start_time = time.time()
        while time.time() - start_time < scroll_time:
            # Scroll down
            driver.execute_script("window.scrollBy(0, 1000);")
            # Pause for a moment to allow the page to load
            time.sleep(0.5)
        page_source = driver.page_source
        newheight = driver.execute_script("return document.body.scrollHeight")
        if newheight == last_height:
            unchanged_iter_count += 1
        else:
            unchanged_iter_count = 0
        print('iters until breaking:', unchanged_iter_count)
        soup = BeautifulSoup(page_source, 'html.parser')
        if unchanged_iter_count > 15:
            break
        time.sleep(random.randint(0,1))
        
    driver.quit()
    soup = BeautifulSoup(page_source, 'html.parser')
    articles = soup.find_all('article', class_='social-item block big')
    p_tags = []

    for article in articles:
        p_tag = article.find('p')
        if p_tag:
            p_tags.append(p_tag.text)

    df = pd.DataFrame(p_tags, columns=['Content'])
    df.to_csv('twitter_data/tweets_'+ team +'.csv', index=False)
```
